```python
import sklearn.linear_model as lm
model = lm.LinearRegression()
```

```python
X_train = [[1],[2],[3]]
y_train = [2.3,3.99,5.4]

fit_model = model.fit(X_train, y_train)
```

```python
X_test = [[3],[5]]
y_test = [6,10]

y_pred = fit_model.predict(X_test)
print(y_pred)
```

```python
from sklearn import metrics

mse = metrics.mean_squared_error(y_test,y_pred)
print(mse)
```

```python
import numpy as np
print('RMSE:',np.sqrt(mse))
```

```python
r2 = metrics.r2_score(y_test,y_pred)
print(r2)
```

```python
fit_model.score(X_test,y_test) #     R2 
```

```python
print(':', fit_model.coef_)
print(':', fit_model.intercept_)
```

```python
boston = sklearn.datasets.load_boston()
print(boston.DESCR)
```

```python
boston.feature_names
```

```python
boston.target[:20]
```

```python
import pandas as pd
df = pd.DataFrame(boston.data,columns=boston.feature_names)
df
```

```python
X = df[['RM']]
y = boston.target
```

```python
# np.array(df['RM'])
# np.array(df[['RM']])
```

```python
from  sklearn import model_selection

X_train, X_test, y_train, y_test = \
model_selection.train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn import linear_model

ex_fit_model = linear_model.LinearRegression().fit(X_train, y_train)
```

```python
y_pred = ex_fit_model.predict(X_test)
```

```python
from sklearn import metrics
mse = metrics.mean_squared_error(y_test,y_pred)
print('MSE :',mse)

r2 = metrics.r2_score(y_test,y_pred) 
# ex_fit_model.score(X_test, y_test)  
print('R2 :',r2)
```

```python
print('coef :',ex_fit_model.coef_)
print('intercept :',ex_fit_model.intercept_)
```
