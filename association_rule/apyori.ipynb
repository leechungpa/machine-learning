```python
tr = [['우유', '계란', '커피'],
      ['빵', '우유', '버터', '휴지', '라면'],
      ['맥주', '땅콩'],
      ['우유', '계란', '빵', '생수'],
      ['햇반', '커피', '간장', '계란']]
```

```python
rules = apyori.apriori(transactions=tr, min_support=0.3, min_confience=0.7)
```

```python
for r in rules :
    print(r)

    '''
    반복자 객체 RelationRecord 반환 결과
    0. items : 빈발 항목 집합
    1. support : 지지도
    2. ordered_statistics : 연관 규칙들의 정보
        - items_base : 선행절
        - items_add : 후행절
        - confidence : 신뢰도
        - lift : 향상도
    '''
    # frozenset : 원소를 바꿀 수 없는 set
```

```python
rules2 = apyori.apriori(tr, min_support=0.3, min_confience=1.2)
```

```python
for rule in list(rules2):
    support = rule[1]
    inf = list(rule[2])
    if len(inf)>1:
        for i in inf:
            print(set(i[0]),'->',set(i[1]),support,i[2],i[3])
```
